## グラフ理論を用いた最適化問題

```python
import networkx as nx
from networkx.algorithms import bipartite
import numpy as np
import itertools
import matplotlib.pyplot as plt
```

```python
group1 = range(5)
group2 = range(5,10)

node_color = ["b"] * 5
node_color.extend(["r"] * 5)
```

```python
g = nx.Graph()
g.add_nodes_from(group1, bipartite=1)
g.add_nodes_from(group2, bipartite=0)
```

```python
for (i,j) in itertools.product(group1, group2):
    val = np.random.randint(1, 10)
    g.add_edge(i, j, weight=val)
```

```python
A,B = bipartite.sets(g)
pos = dict()
pos.update((n,(1,i)) for i,n in enumerate(A))
pos.update((n,(2,i)) for i,n in enumerate(B))
```

```python
edge_width = [ d['weight']*0.3 for (u,v,d) in g.edges(data=True)]

nx.draw_networkx(g, pos, node_color=node_color)
nx.draw_networkx_edges(g, pos, width=edge_width)
plt.axis("off")
plt.show()
```

```python
d = nx.max_weight_matching(g)

for (i, j) in list(g.edges()):
    if (i, j) not in d:
        if (j, i) not in d:
            g.remove_edge(i,j)
```

```python
def _can_apply(order, product):
    """紐付可否判定
    
    Args:
        order (Series)  : 注文データ
        product (Series): 商品データ
    
    Returns:
        Bool            : True：紐付可、False：紐付不可
    """
    if order["grade"] == product["grade"] and \
        order["width_min"] <= product["width"] and \
        order["width_max"] >= product["width"]:
        
        return True
    
    return False
```

```python
def applicable_quantity(order, product):
    """紐付量計算
    
    Args:
        order (Series)  : 注文データ
        product (Series): 商品データ
    
    Returns:
        int             : 紐付量
    """
    if _can_apply(order, product):
        
        return min(order["weight"], product["weight"])
    
    return 0
```

```python
# 貪欲法

greedy_aq = dict()    #商品に対する紐付量を保持する辞書を定義
greedy_match = dict() #商品に対する紐付注文IDを保持する辞書を定義

for oid, order in orders.iterrows(): #注文ノードを先頭から見ていく
    for sid, product in products.iterrows(): #商品ノードを先頭から見ていく
        if sid in greedy_aq and greedy_aq[sid] > 0: continue #greedy_aq内の該当するキーの値が0より大きい（すでに紐付いている）とき、次の商品IDを見る
        aq = applicable_quantity(order, product) #紐付け可能な時、紐付け量aqは紐付量計算サブルーチンの戻り値
        if aq > 0:
            greedy_aq[sid] = aq   #商品に対する紐付量を保持する辞書に紐付け量を追加
            greedy_match[sid] = oid   #商品に対する紐付注文IDを保持する辞書をに注文IDを追加
            break
```

```python
# 二部グラフの最大マッチング問題

G = nx.Graph() #空の無向グラフオブジェクトを作る

for oid in orders.index:
    G.add_node(oid, bipartite=0) #注文IDノードを作る

for sid in products.index:
    G.add_node(sid, bipartite=1) #商品IDノードを作る
    
for oid, order in orders.iterrows():
    for sid, product in products.iterrows():
        aq = applicable_quantity(order, product)
        if aq > 0:
            G.add_edge(oid, sid, weight=aq) #紐付け量>0ならば、その紐付け量を重みとするエッジを張る
```

```python
# 重み付き最大マッチングの実行
match = nx.max_weight_matching(G)

# for (i,j) in list(G.edges()):
#     if (i, j) not in d:
#         if (j, i) not in d:
#             G.remove_edge(i,j)

# edge_width = [ d['weight']*0.2 for (u,v,d) in G.edges(data=True)]

opt_aq = dict()    #商品に対する紐付量を保持する辞書を定義
opt_match = dict() #商品に対する紐付注文IDを保持する辞書を定義

for sid, product in products.iterrows():
    opt_match = oid = match   #商品に対する紐付量を保持する辞書に紐付け量を追加
    #weight = product['weight']
    print(G.edges.data())
    print(opt_aq[sid])
    #商品に対する紐付注文IDを保持する辞書をに注文IDを追加
    #opt_aq[sid] = G.edges(oid, sid, weight=weight)  
```
